Exemplo:

```
num = input('Por favor digite um número ')
num = int(num) #int é para garantir que o num seja um número inteiro.
                 #float é para garantir que todos os outros tipos sem ser inteiro.
if num > 5 :
  print(f'O número inserido é maior do que 5 e é {num}') #Formatted String - string formatado
elif num < 5 :
  print(f'O número inserido é menor do que 5 e é {num}')
else :
  print('O número inserido é 5.')
```

```
num = input('Por favor digite um número ')
num = int(num) #int é para garantir que o num seja um número inteiro.
                 #float é para garantir que todos os outros tipos sem ser inteiro.
if num > 5 :
  print('O número inserido é maior do que 5 e é ' + str(num))
elif num < 5 :
  print('O número inserido é menor do que 5 e é ' , num)
else :
  print('O número inserido é 5.')
```

Exercício para classe: Pega dois números da entrada e printa o maoir e se forem iguais printa"São iguais."

```
n1 = input('Insira um número ')
n1 = int(n1)
n2 = input('Insira mais um número ')
n2 = int(n2)

if n1 > n2 :
  print('O primeiro número é o maior.')
elif n2 > n1 :
  print('O segundo número é o maior.')
else : 
  print('Os números são iguais.')
```

##While loop

```
i = 0

while i < 10 : 
  print(i)
  i = i + 1
```

```
i = 0

while i <= 10 : 
  print(i)
  i = i + 1
```

```
i = 0
while True : #Condição infinita
  if i < 10 :
    print(i)
    i = i + 1
  else :
    break #Com break a gente quebra qualquer loop
```

##List comprehension

mecanismo para gerar lista com loop implícito

```
#Gerar lista com 15 elementos 
```

```
l = []

for i in range(15) :
  l.append(i**2)
```

```
l
```

```
l_com = [i**2 for i in range(15)]
```
