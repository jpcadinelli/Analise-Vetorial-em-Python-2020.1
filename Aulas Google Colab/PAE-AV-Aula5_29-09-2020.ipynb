##Multiple plots

```
from plotly.subplots import make_subplots

t = np.linspace(-2*np.pi, 2*np.pi,100)
x = np.sin(t)
y = np.cos(t)

fig = make_subplots(rows = 1, cols = 2)

fig.add_scatter(x = t, y = x, row = 1, col = 1, name = 'sin')
fig.add_scatter(x = t, y = y, row = 1, col = 2, name = 'cos')

fig.show()
```

```
fig.update_layout(title = 'Dois Gráficos')
```

##Curves

```
fig = go.Figure()

t = np.linspace(-2*np.pi, 2*np.pi,100)
x = np.sin(t)
y = np.cos(t)
z = t

fig.add_scatter3d(x = x, y = y, z = z, mode = 'lines')

fig.update_layout(title = 'Helice',
    scene = dict(
    xaxis = dict(range = [-5,10], title = 'eixo x'),
    yaxis = dict(range = [-5,10], title = 'eixo y'),
    zaxis = dict(range = [-5,10], title = 'eixo z')),
    margin = dict(r = 20, l = 10, t = 30, b = 20))
```
