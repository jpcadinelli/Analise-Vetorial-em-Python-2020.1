```
import sympy as sp
import numpy as np
import sympy.vector as sv
```

```
sp.init_printing()
R = sv.CoordSys3D('R')
def helice(t):
  return sp.cos(t)*R.i + sp.sin(t)*R.j + t*R.k
t = sp.symbols('t', real=True)
```

Monte uma funo que pegue uma curva e retorne ostres vetores $\space \vec T(t), \space \vec N(t) \space e \space \vec B(t)$

```
def UB(curve):
    der = curve.diff()
    mod = der.magnitude().simplify()
    tg = der/mod
    der_tg = tg.diff()
    mod_tg = der_tg.magnitude().simplify()
    normal = der_tg/mod_tg
    binormal = tg.cross(normal)
    return binormal,normal,tg
```

```
frenet = UB(helice(t))
```
