```
def gradiente(func, coordinate=None):
  if coordinate:
    R = coordinate
  else: 
    R = sv.CoordSys3D("R") 
    vars = list(sp.ordered(func.free_symbols))
    grad = func.diff(vars[0]) * R.i + func.diff(vars[1]) * R.j
  return grad  
```

```
gradiente(scalar(x,y))
```

```
C = sv.CoordSys3D('C')
gradiente(scalar(x,y), coordinate=C)
```

```
av.plot_vector_field(gradiente(scalar(x,y)), (x,-5,5), (y,-5,5))
```

```
#Marco
def gradienteMarco(func, vars, coordinate=None):
    if coordinate:
        R = coordinate
    else:
        R = sv.CoordSys3D("R")
    if len(vars) == 1:
        grad = func.diff(vars[0])*R.i
    if len(vars) == 2:
        grad = func.diff(vars[0])*R.i + func.diff(vars[1]) * R.j
    elif len(vars) == 3:
        grad = func.diff(vars[0])*R.i + func.diff(vars[1]) * R.j + func.diff(vars[2])*R.k

    return grad
```
