```
l1 = [x for x in range(21)]
l1
```

##Funções

$f(x)=x^2+3$

```
def func(x) :
  return x**2 + 3
```

```
func(3)
```

$f(x,y)=x^2+y^2$

```
def func1(x,y) :
  return x**2 + y**2

func1(2,3)
```

```
def func_1(x,y=1) :
  return x**2 + y**2

func_1(2)
```

```
def func_2(x=None,y=None) :
  if (x==None) or (y==None) :
    print("Esqueceu de uma variável.")
  else :
   return x**2 + y**2
```

```
func_2(4)
```

EX2: 

Escreva uma função que pega o nome do usuário e se esse for 'Rebecca' printa 'Tudo bem Rebecca' e se n for Rebecca, preinta 'Não te conheço'. Dica: use print em vez de return

```
def rebeccas_dog(name) :
  if name=='Rebecca' :
    print('Tudo bem Rebecca!')
  else :
    print('Não te conheço!')

name = input('Qual seu nome?  ')
rebeccas_dog(name)
```

```
def rebeccas_dog(name) :
  if name.lower()=='rebecca' :  #lower muda tudo para minúsculo
    print('Tudo bem Rebecca!')
  else :
    print('Não te conheço!')

name = input('Qual seu nome?  ')
rebeccas_dog(name)
```
