```
import sympy as sp
import numpy as np
import sympy.vector as sv
```

```
R = sv.CoordSys3D("R")
```

```
sp.init_printing()
```

```
t,s,u,x,y,z= sp.symbols("t,s,u,x,y,z",real = True )
```

##Reparametrização

$s = S(t) = \int_a^t |r'(u)|du \rightarrow t =S^{-1}(s)$

```
def helice(t):
  return sp.cos(t)*R.i +sp.sin(t)*R.j +t*R.k
```

```
helice(t)
```

```
helice(t).dot(R.i).subs(t,3)
```

```
rapidez = helice(t).diff().magnitude()   #magnitude é pra calcular o modulo da função vetorial
```

```
rapidez
```

```
rapidez = helice(t).diff().magnitude().simplify() 
rapidez
```

```
def S(t):
  return sp.integrate(rapidez,(u,0,t))
```

```
S(t)
```

```
eq = sp.Eq(s,S(t))
```

```
eq
```

```
sol = sp.solve(eq,t,dict = True)
```

```
sol[0][t]
```

```
helice(t).subs(t,sol[0][t])
```

##Curvatura
$\kappa = \frac{d \vec T}{ds} = \frac{|\vec{T'}(t)|}{|\vec{r'}(t)|}$

$\vec{T}{(t)} = \frac{\vec{r'}{t}}{|{\vec{r'}{t}}|}$

Algumas coisas são igual nadar na piscina de mel, vc n anda pra frente, mas é gostoso

```
def UT(curve):
  der = curve.diff().simplify()
  mod = der.magnitude().simplify()
  return (der/mod).simplify()
```

```
UT(helice(t))
```

```
UT(helice(t)).magnitude().simplify() #vetor tangente unitário
```

```
def circle(t, a=1):
  return a*sp.cos(t)*R.i + a*sp.sin(t)*R.j
```

```
cir = circle(t,5)
```

```
cir
```

```
t_linha = UT(cir).diff().magnitude()
t_linha
```

```
t_linha = UT(cir).diff().magnitude().simplify()
t_linha
```

```
cir_linha = cir.diff().magnitude().simplify()
```

```
cir_linha
```

```
kappa = t_linha/cir_linha
```

```
kappa
```

```
def curvatura(curva):
    der = curva.diff()
    mod = der.magnitude().simplify()
    tg = der/mod
    mod_tg = tg.magnitude()
    k = mod_tg / mod 
    return k.simplify()
```

```
curvatura(cir)
```
