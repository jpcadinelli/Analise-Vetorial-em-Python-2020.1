```
import numpy as np
import sympy as sp
import sympy.vector as sv

sp.init_printing()
R = sv.CoordSys3D('R')
x, y, z, t = sp.symbols('x, y, z, t', real = True)

def integral_de_linha(func, a, densidade, b):
  densidade_np = sp.lambdify((b[0],b[1],b[2]), densidade, 'numpy')
  dr = func.diff()
  mod = dr.magnitude().simplify()

  ft = tuple(func.components.values())
  paramt = densidade_np(ft[0], ft[1], ft[2])

  integral = paramt * mod
  return float(sp.integrate(integral, (a[0],a[1],a[2])))

#Teste
f = t*R.i + t**3*R.j + t**2*R.k
a = (t,0,1)
d = 2*x + 4*y
b = (x, y, z)
integral_de_linha(f, a, d, b)  
```

###Integrla de linha de campo vetorial

```
import numpy as np
import sympy as sp
import sympy.vector as sv

sp.init_printing()
R = sv.CoordSys3D('R')
x,y,z,t = sp.symbols('x y z t',real=True)

def integral_de_linha_vet(curve1, a, field1, b):
  ft = (field1.components.values())
  ct = tuple(curve1.components.values())
  paramet_field = field1.subs(b[0],ct[0]).subs(b[1],ct[1]).subs(b[2],ct[2])
  dr = curve1.diff()
  integrand = paramet_field.dot(dr)
  return sp.integrate(integrand, (a[0], a[1], a[2]))

#Teste
curva = t*R.i + sp.sin(t)*R.j + R.k
c_a = (t,0,sp.pi)
d = z**3*R.i + y**2*R.j + x*R.k
d_b = (x,y,z)  
integral_de_linha_vet(curva, c_a, d, d_b)
```
