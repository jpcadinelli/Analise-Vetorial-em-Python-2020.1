## Solvers

```
eq = sp.Eq(x**2,1)

eq
```

```
sp.solve(eq)
```

```
eq1 = sp.Eq(sp.sin(x)-1,x)

eq1
```

```
sp.solveset(eq1,x)
```

```
eq1 = sp.Eq(x+y+z,1)
eq2 = sp.Eq(x+y+2*z,-3)
eq3 = sp.Eq(2*x + y-z,-10)

sp.pprint(eq1)
sp.pprint(eq2)
sp.pprint(eq3)
```

```
sol = sp.solve([eq1,eq2,eq3], (x,y,z))
```

```
sol[x]
```

```
sol
```

```
eq1 = sp.Eq(x+y**2+z,1)
eq2 = sp.Eq(x+y+2*z,-3)
eq3 = sp.Eq(2*x + y-z,-10)

display(eq1)
display(eq2)
display(eq3)
```

```
sp.solve([eq1,eq2,eq3], (x,y,z))
```
